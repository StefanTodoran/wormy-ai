```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set_context()
```

```python
data = {
    "Category": ["Friendly", "Job Search", "Marketing", "Reminder"],
    "Percentage": [100, 99, 2, 3]
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x="Category", y="Percentage", data=df, palette="pastel")
plt.ylim(0, 100)  # Assuming percentages range from 0 to 100
plt.title("Effect of Email Category on Reply Rate")
plt.xlabel("Email Category")
plt.ylabel("Average Reply Rate")
plt.show()
```
